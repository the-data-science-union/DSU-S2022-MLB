```python
import numpy as np
import pandas as pd
```

```python
Outfield = pd.read_csv('Outfield.csv')
```

```python
Outfield.head()
```

```python
MLOutfield = Outfield.drop(['year', 'Tm', 'G'], axis = 1)
MLOutfield.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
X = MLOutfield.drop('Playoff', axis = 1) 
test_ml = pd.series(np.where(MLOutfield.Playoff == 'Yes', 1, 0))
Y = MLOutfield.Playoff 
Xtrain, Xtest, Ytrain, Ytest = train_test_split(X, Y)
```

```python
Xtrain.shape # checking sizes
```

```python
Xtest.shape
```

```python
Ytrain.shape
```

```python
Ytest.shape
```

```python
logreg_model = LogisticRegression(max_iter = 10000)
logreg_model.fit(Xtrain, Ytrain)
predicted = logreg_model.predict(Xtest)
actual = Ytest
print(metrics.accuracy_score(predicted, actual))
```
